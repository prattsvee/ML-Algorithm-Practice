```python
import numpy as np
import matplotlib.pyplot as plt

num_trials = 6
target_mean = 0.5
def is_difference_exceeded(epsilon):
    sample_a = np.random.choice([0, 1], num_trials, replace=True)
    mean_a = np.mean(sample_a)

    sample_b = np.random.choice([0, 1], num_trials, replace=True)
    mean_b = np.mean(sample_b)

    max_difference = max(abs(mean_a - target_mean), abs(mean_b - target_mean))
    return max_difference > epsilon

def calculate_probability(epsilon):
    return np.mean([is_difference_exceeded(epsilon) for _ in range(10000)])

vectorized_probability = np.vectorize(calculate_probability)
epsilon_values = np.linspace(0, 1, 100)
probability_values = vectorized_probability(epsilon_values)
plt.plot(epsilon_values, probability_values, label='Probability')
plt.plot(epsilon_values, 4 * np.exp(-12 * epsilon_values**2), label='Reference', color='green')
plt.xlabel('Epsilon')
plt.ylabel('Probability')
plt.legend()
plt.show()
```
