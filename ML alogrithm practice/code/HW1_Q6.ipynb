```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

num_points = 20
x_values = np.linspace(-20, 20, num_points)
y_values = x_values + np.random.uniform(-50, 50, num_points)
above_line = y_values > x_values
below_line = ~above_line

labels = np.ones(num_points)
labels[below_line] = -1

plt.scatter(x_values[above_line], y_values[above_line], marker='o', color='red', label='+1 (Above Line)')
plt.scatter(x_values[below_line], y_values[below_line], marker='o', color='blue', label='-1 (Below Line)')
plt.plot(x_values, x_values, linestyle='-', color='black', label='Decision Boundary (y=x)')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Linearly Separable Data with Labels')
plt.grid(True)
plt.show()

weights = np.random.rand(3)
c = 0
for epoch in range(100):
    for inputs, label in zip(np.column_stack((x_values, y_values)), labels):
        prediction = predict(inputs, weights)
        weights += 0.01 * (label - prediction) * np.concatenate([np.array([1]), inputs])
        if label - prediction != 0:
          c += 1

x_decision_boundary = np.linspace(-20, 20, 100)
y_decision_boundary = -(weights[0] + weights[1] * x_decision_boundary) / weights[2]
plt.plot(x_decision_boundary, y_decision_boundary, linestyle='-', color='green', label='Learned Decision Boundary')

plt.scatter(x_values[above_line], y_values[above_line], marker='o', color='red', label='+1 (Above Line)')
plt.scatter(x_values[below_line], y_values[below_line], marker='o', color='blue', label='-1 (Below Line)')
plt.plot(x_values, x_values, linestyle='-', color='black', label='Decision Boundary (y=x)')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Perceptron Learning Algorithm')
plt.grid(True)
plt.show()

print("Converging at iteration : ",c)

def predict(inputs, weights):
    inputs_with_bias = np.concatenate([np.array([1]), inputs])
    activation = np.dot(weights, inputs_with_bias)
    return 1 if activation > 0 else -1
```
